```python
run=126
ds_str = 'exp=cxilr6716:run=%d:smd' % run
ds = DataSource(ds_str)
det = Detector('CxiDs2.0:Cspad.0')
```

```python
events=ds.events()
max_evt=200
imgs=[]
for nevent,evt in enumerate(events):
    if nevent>=max_evt: 
        break
    try:
        img = det.image(evt)
    except ValueError:
        break
    imgs.append(img)
imgs=np.array(imgs)
```

```python
plt.figure(figsize=[8,8])
plt.imshow(imgs.mean(0)[300:1300,300:1300],aspect=1,vmin=50,vmax=300, cmap='hot')
plt.colorbar()
# np.save('sb_img.npy',imgs.mean(0))
```
