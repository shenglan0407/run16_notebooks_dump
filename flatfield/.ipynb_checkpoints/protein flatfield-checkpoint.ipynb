```python
def flatten_and_mask_shots(shots, mask,):
    num_shots = shots.shape[0]
    size = shots.shape[1]*shots.shape[2]
    flat_mask = mask.reshape(size)
    flat_shots =  shots.reshape( (num_shots,size))*flat_mask
    
    return flat_shots[:,flat_mask]

def unflatten_shots(flat_shots,mask):
    num_shots = flat_shots.shape[0]
    shape = mask.shape
    flat_mask = mask.reshape(shape[0]*shape[1]
                            )
    shots = np.zeros( (num_shots,shape[0]*shape[1]), dtype = flat_shots.dtype)
    shots[:, flat_mask ] = flat_shots
    
    return shots.reshape( (num_shots,shape[0],shape[1]))


def diff(a,b):
    return (a/a.mean())/(b/b.mean())-1
```

```python
# difference with reference shot
def diff_mask(x,ref,mask):
    mean_int=x[mask].mean()
    ref_int = ref[mask].mean()
    
    return np.nan_to_num((x/mean_int)/(ref/ref_int)-1)*pmask
```

```python
diffs=[]
cali_diffs=[]
ref_ind=2
for ii in range(ave_shots_interp.shape[0]):
    if ii != ref_ind:
        diffs.append(diff_mask(ave_shots_interp[ii], ave_shots_interp[ref_ind],pmask))
        cali_diffs.append(diff_mask(cali_ave_shots_interp[ii], cali_ave_shots_interp[ref_ind],pmask))
diffs=np.array(diffs)
cali_diffs= np.array(cali_diffs)
```
